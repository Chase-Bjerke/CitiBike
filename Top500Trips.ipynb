# Top 500 Trips

## Libraries/Imports

```python
import pandas as pd
import os
```

```python
# Import merged csv
top500_df = pd.read_csv("02_Data/Prepared_Data/top500_routes.csv")
```

```python
top500_df.head()
```

```python
# Add bins for the arc stroke size for the kelpergl map
def stroke_bin(trip):
    if trip < 1800:
        return 0.2
    elif trip < 2200:
        return 1.0
    elif trip < 3000:
        return 2.0
    else:
        return 3.0

top500_df["stroke_width"] = top500_df["trip_count"].apply(stroke_bin)
```
